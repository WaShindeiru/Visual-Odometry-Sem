```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
%matplotlib notebook
%matplotlib inline
```

```python
path = "/media/washindeiru/Hard Disc/odometry_files/odometry_files/m2dgr/data"
# image = "1628059159326742649.png"
image = "1628058907164302588.png"
```

```python
img = cv2.imread(path + "/" + image, cv2.IMREAD_GRAYSCALE)
plt.imshow(img, cmap='gray')
plt.show()
```

```python
print(img.shape)
```

```python
w, h = img.shape
print(w, h)
```

```python
camera_matrix = np.array([[540.645056202188, 0., 626.4125666883942], 
        [0., 539.8545023658869, 523.947634226782], 
        [0., 0., 1.]])

distortion_coefficients = np.array([-0.07015146608431883, 0.008586142263125124, -0.021968993685891842, 0.007442211946112636])
```

```python
newcameramatrix, roi = cv2.getOptimalNewCameraMatrix(camera_matrix, distortion_coefficients, (w, h), 1, (w, h))
```

```python
dst = cv2.undistort(img, camera_matrix, distortion_coefficients, None, newcameramatrix)
# 
# x, y, w, h = roi
# dst = dst[y:y+h, x:x+w]

plt.imshow(dst, cmap='gray')
plt.show()
```

```python
cv2.imwrite('calibresult.png', dst)
```
